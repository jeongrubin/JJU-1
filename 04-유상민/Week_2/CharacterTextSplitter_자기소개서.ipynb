```python
# data/appendix-keywords.txt 파일을 열어서 f라는 파일 객채를 생성합니다
with open('/content/data/appendix-keywords.txt') as f:
  file_1 = f.read() # 파일의 내용을 읽어서 file 변수에 저장합니다.

with open('/content/data/자기소개서_유상민.txt') as f:
  file_2 = f.read() # 파일의 내용을 읽어서 file 변수에 저장합니다.
```

```python
print(file_2[:100])
```

```python
from langchain_text_splitters import CharacterTextSplitter

# CharacterTextSplitter를 사용하여 텍스트를 청크(chunk)로 분할하는 코드
text_splitter = CharacterTextSplitter(
    # 텍스트를 분할할 때 사용할 구분자를 지정합니다. 기본값은 "\n\n"입니다
    separator = '\n\n',
    # 분할된 텍스트 청크의 최대 크기를 지정합니다 (문자 수)
    chunk_size=210,
    # 분할된 텍스트 청크 간의 중복되는 문자 수를 지정합니다
    chunk_overlap=0,
    # 텍스트의 길이를 계산하는 함수를 지정합니다
    length_function=len,
)
```

```python
texts = text_splitter.create_documents([file_1, file_2])
print(len(texts[0].page_content))
print(texts[32])
```

```python
metadatas = [
    {'document': 1},
    {'document': 2},
]
documents = text_splitter.create_documents(
    [
        file_1,
        file_2,
    ],
    metadatas=metadatas,
)

print(documents[32])
```
