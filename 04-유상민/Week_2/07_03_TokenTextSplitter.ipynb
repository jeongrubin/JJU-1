# TokenTextSplitter

언어 모델에는 토큰 제한이 있습니다. 따라서 토큰 제한을 초과하지 않아야 합니다.

TokenTextSplitter는 텍스트를 토큰 수를 기반으로 청크를 생성할 때 유용합니다

```python
# data/appendix-keywords.txt 파일을 열어서 f라는 파일 객채를 생성합니다
with open('/content/data/appendix-keywords.txt') as f:
  file = f.read() # 파일의 내용을 읽어서 file 변수에 저장합니다.
```

```python
print(file[:500])
```

```python
!pip install -q tiktoken
```

```python
from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter.from_tiktoken_encoder(
    chunk_size = 300,
    chunk_overlap = 0,

)

texts = text_splitter.split_text(file)
```

```python
print(len(texts)) # 분할된 청크의 개수를 출력
```

```python
# texts 리스트의 첫 번째 요소를 출력
print(texts[1])
```

```python
# texts 리스트의 두 번째 요소를 출력
print(texts[2])
```

```python
from langchain.text_splitter import TokenTextSplitter

text_splitter = TokenTextSplitter(
    chunk_size = 200,
    chunk_overlap = 0,

)

texts = text_splitter.split_text(file)
print(texts[0])
```

# SentenceTransformers

```python
from langchain.text_splitter import SentenceTransformersTokenTextSplitter


splitter = SentenceTransformersTokenTextSplitter(chunk_size = 200, chunk_overlap=0)
```

```python
# data/appendix-keywords.txt 파일을 열어서 f라는 파일 객채를 생성합니다
with open('/content/data/appendix-keywords.txt') as f:
  file = f.read() # 파일의 내용을 읽어서 file 변수에 저장합니다.

print(file[:350])
```

```python
count_start_and_stop_tokens = 2

text_token_count = splitter.count_tokens(text=file) - count_start_and_stop_tokens
print(text_token_count)
```

```python
text_chunks = splitter.split_text(text=file) # 텍스트를 청크로 분
```

```python
print(text_chunks[0])
```

# Hugging Face tokenizer

```python
from transformers import GPT2Tokenizer

hf_tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
```

```python
# data/appendix-keywords.txt 파일을 열어서 f라는 파일 객채를 생성합니다
with open('/content/data/appendix-keywords.txt') as f:
  file = f.read() # 파일의 내용을 읽어서 file 변수에 저장합니다.

print(file[:350])
```

```python
text_splitter = CharacterTextSplitter.from_huggingface_tokenizer(
    hf_tokenizer,
    chunk_size = 300,
    chunk_overlap = 50,
)

texts = text_splitter.split_text(file)
```
