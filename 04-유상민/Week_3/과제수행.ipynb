```python
from langchain.document_loaders import PyPDFLoader, PyMuPDFLoader, PDFMinerLoader
from langchain_community.document_loaders import PyPDFium2Loader
import pandas as pd

def extract_text_from_pdf(file_path: str):
    """PDF 파일에서 각 로더별로 500자씩 텍스트를 추출."""
    loaders = {
        "PyPDFLoader": PyPDFLoader,
        "PyMuPDFLoader": PyMuPDFLoader,
        "PyPDFium2Loader": PyPDFium2Loader,
        "PDFMinerLoader": PDFMinerLoader
    }

    extracted_texts = []

    for loader_name, loader_class in loaders.items():
        try:
            loader_instance = loader_class(file_path)
            docs = loader_instance.load()
            text = docs[0].page_content[:500]  # 첫 페이지에서 500자 추출
            extracted_texts.append({"모델명": loader_name, "텍스트": text})
        except Exception as e:
            extracted_texts.append({"모델명": loader_name, "텍스트": f"오류 발생: {str(e)}"})

    return pd.DataFrame(extracted_texts)

if __name__ == "__main__":
    file_path = "/content/data/소상공인_전기요금_특별지원_사업_시행_수정공고(제2024-557호).pdf"

    extracted_df = extract_text_from_pdf(file_path)

    print("\n=== 각 로더별 추출 텍스트 ===")
    for _, row in extracted_df.iterrows():
        print(f"\n[{row['모델명']}]")
        print(row['텍스트'])
        print('-' * 80)
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.chat_models import ChatOpenAI
from langchain.document_loaders import PyPDFLoader, PyMuPDFLoader, PDFMinerLoader
from langchain_community.document_loaders import PyPDFium2Loader
import re
from typing import Dict
import pandas as pd

def evaluate_text(text: str) -> Dict:
    """텍스트에 대한 정량적 평가 수행 (0-5점 척도)"""
    # 1. 텍스트 길이 평가 (0-5점)
    text_length = len(text)
    length_score = min(5, int(text_length / 100))

    # 2. 문장 구조 평가 (0-5점)
    sentence_count = len(re.findall(r'[.!?]', text))
    sentence_score = min(5, int(sentence_count / 2))

    # 3. 단어 다양성 평가 (0-5점)
    unique_words = len(set(text.split()))
    word_diversity_score = min(5, int(unique_words / 20))

    # 4. 특수문자 적절성 평가 (0-5점)
    special_chars = len(re.findall(r'[^a-zA-Z0-9\s]', text))
    special_char_score = min(5, 5 - min(5, int(special_chars / 50)))

    # 5. 가독성 평가 (0-5점)
    words = len(text.split())
    avg_words_per_sentence = words / max(1, sentence_count)
    readability_score = min(5, 5 - abs(15 - avg_words_per_sentence) // 3)

    total_score = length_score + sentence_score + word_diversity_score + special_char_score + readability_score

    return {
        "텍스트 길이": length_score,
        "문장 구조": sentence_score,
        "단어 다양성": word_diversity_score,
        "특수문자 적절성": special_char_score,
        "가독성": readability_score,
        "총점": total_score
    }

def process_pdf(file_path: str):
    model = ChatOpenAI(
        model="gpt-4o",
        max_tokens=2048,
        temperature=0.1,
    )

    prompt = PromptTemplate(
        input_variables=["content", "score"],
        template="다음은 PDF 텍스트입니다:\n\n{content}\n\n총점은 {score}점입니다(25점 만점). 이 점수를 기준으로 간단한 피드백을 작성하세요."
    )

    chain = LLMChain(llm=model, prompt=prompt)

    loaders = {
        "PyPDFLoader": PyPDFLoader,
        "PyMuPDFLoader": PyMuPDFLoader,
        "PyPDFium2Loader": PyPDFium2Loader,
        "PDFMinerLoader": PDFMinerLoader
    }

    evaluation_results = []
    feedback_results = []

    for loader_name, loader_class in loaders.items():
        try:
            loader_instance = loader_class(file_path)
            docs = loader_instance.load()
            text = docs[0].page_content[:500]

            evaluation = evaluate_text(text)
            evaluation["모델명"] = loader_name
            evaluation_results.append(evaluation)

            feedback = chain.run({
                "content": text,
                "score": evaluation["총점"]
            })

            feedback_results.append({
                "모델명": loader_name,
                "피드백": feedback
            })

        except Exception as e:
            evaluation_results.append({
                "모델명": loader_name,
                "텍스트 길이": 0,
                "문장 구조": 0,
                "단어 다양성": 0,
                "특수문자 적절성": 0,
                "가독성": 0,
                "총점": 0
            })
            feedback_results.append({
                "모델명": loader_name,
                "피드백": f"오류 발생: {str(e)}"
            })

    return pd.DataFrame(evaluation_results), pd.DataFrame(feedback_results)

if __name__ == "__main__":
    # pandas 출력 옵션 설정
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)
    pd.set_option('display.max_colwidth', 20)
    pd.set_option('display.colheader_justify', 'center')

    file_path = "/content/data/소상공인_전기요금_특별지원_사업_시행_수정공고(제2024-557호).pdf"

    metrics_df, feedback_df = process_pdf(file_path)

    print("\n=== 정량 평가 결과 ===")
    print("\n[평가 기준]")
    print("- 텍스트 길이: 추출된 텍스트의 적절한 길이 (0-5점)")
    print("- 문장 구조: 문장의 완성도 (0-5점)")
    print("- 단어 다양성: 사용된 단어의 다양성 (0-5점)")
    print("- 특수문자 적절성: 특수문자 사용의 적절성 (0-5점)")
    print("- 가독성: 문장당 단어 수의 적절성 (0-5점)")
    print("- 총점: 전체 평가 점수 (0-25점)\n")

    # 정량 평가 결과 출력 (형식 지정)
    sorted_metrics = metrics_df.sort_values(by="총점", ascending=False)

    # 각 열의 최대 길이 계산
    col_widths = {}
    for col in sorted_metrics.columns:
        max_len = max(
            len(str(col)),
            sorted_metrics[col].astype(str).map(len).max()
        )
        # 열 폭에 여유를 두기 위해 +2 정도 추가
        col_widths[col] = max_len + 2

    # 헤더 생성
    header_parts = [f"{col:^{col_widths[col]}}" for col in sorted_metrics.columns]
    header_line = " | ".join(header_parts)

    # 구분선(가운데 정렬된 헤더와 동일한 길이를 가지도록) 생성
    separator_parts = ["-" * col_widths[col] for col in sorted_metrics.columns]
    # 구분선 사이에 "+", "-", "|", "─" 등 원하는 문자를 활용 가능
    separator_line = "-+-".join(separator_parts)

    # 데이터 행 생성
    rows = []
    for _, row in sorted_metrics.iterrows():
        row_parts = [f"{str(val):^{col_widths[col]}}" for col, val in row.items()]
        formatted_row = " | ".join(row_parts)
        rows.append(formatted_row)

    # 테이블 출력
    print(header_line)
    print(separator_line)
    for row in rows:
        print(row)

    # 최고 성능 모델 출력
    best_model = sorted_metrics.iloc[0]
    print(f"\n최고 성능 모델: {best_model['모델명']} (총점: {best_model['총점']}/25점)")


    # 피드백 결과 출력
    print("\n=== 상세 피드백 ===")
    for _, row in feedback_df.iterrows():
        print(f"\n[{row['모델명']}]")
        print(row['피드백'])
        print('-' * 80)
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.chat_models import ChatOpenAI
from langchain.document_loaders import PyPDFLoader, PyMuPDFLoader, PDFMinerLoader
from langchain_community.document_loaders import PyPDFium2Loader

def extract_text_from_pdf(file_path: str):
    """PDF에서 텍스트 추출"""
    loaders = {
        "PyPDFLoader": PyPDFLoader,
        "PyMuPDFLoader": PyMuPDFLoader,
        "PyPDFium2Loader": PyPDFium2Loader,
        "PDFMinerLoader": PDFMinerLoader
    }

    extracted_texts = []
    for loader_name, loader_class in loaders.items():
        try:
            loader_instance = loader_class(file_path)
            docs = loader_instance.load()
            text = docs[0].page_content[:500]  # 500자 추출
            extracted_texts.append({"모델명": loader_name, "텍스트": text})
        except Exception as e:
            extracted_texts.append({"모델명": loader_name, "텍스트": f"오류 발생: {str(e)}"})

    return extracted_texts


def analyze_with_gpt(texts):
    """GPT를 사용해 정량 평가표 및 피드백 생성"""
    model = ChatOpenAI(
        model="gpt-4o",
        max_tokens=2048,
        temperature=0.1,
    )

    prompt_template = PromptTemplate(
        input_variables=["texts"],
        template=(
            "다음은 여러 PDF 추출 모델에서 가져온 텍스트들입니다:\n\n{texts}\n\n"
            "각 텍스트에 대해 다음을 수행하세요:\n"
            "1. 정량 평가표를 작성 (항목: 텍스트 길이, 문장 구조, 단어 다양성, 특수문자 적절성, 가독성, 총점). "
            "표는 아래 형식으로 작성하세요:\n"
            "| 모델명          | 텍스트 길이 | 문장 구조 | 단어 다양성 | 특수문자 적절성 | 가독성 | 총점 |\n"
            "|-----------------|-------------|-----------|-------------|-----------------|--------|------|\n"
            "\n"
            "2. 각 모델별 피드백을 작성. 피드백은 별도의 목록으로 작성하세요.\n"
            "결과는 Markdown 형식으로 반환하세요."
        )
    )

    chain = LLMChain(llm=model, prompt=prompt_template)

    result = chain.run({"texts": texts})
    return result


if __name__ == "__main__":
    # PDF 파일 경로
    file_path = FILE_PATH

    # PDF 텍스트 추출
    extracted_texts = extract_text_from_pdf(file_path)

    # 모델별 텍스트 출력
    print("\n=== 각 모델별 추출된 텍스트 ===")
    for item in extracted_texts:
        print(f"\n[{item['모델명']}]")
        print(item['텍스트'])
        print('-' * 80)

    # GPT 입력 준비
    gpt_input = "\n".join(
        f"모델명: {item['모델명']}\n텍스트: {item['텍스트']}"
        for item in extracted_texts
    )

    # GPT 분석 실행
    gpt_result = analyze_with_gpt(gpt_input)

    # 결과 출력
    print("\n=== GPT 분석 결과 ===")
    print(gpt_result)
```
