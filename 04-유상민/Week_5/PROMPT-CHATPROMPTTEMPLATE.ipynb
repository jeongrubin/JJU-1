```python
import os
import time
from datetime import datetime
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

# ChatPromptTemplate 설정
chat_prompt = ChatPromptTemplate.from_template(
    "{nation}의 현재시간은 {year}년, {month}월, {day}일, {hour}시, {minute}분, {second}초입니다."
)

# LLM 설정
llm = ChatOpenAI(
    model='gpt-4o',
    temperature=0,
)

# 10번 반복하여 랜덤 국가 및 시간 출력
for _ in range(10):
    # 각 반복에서 새로운 랜덤 국가 생성
    nation_prompt = ChatPromptTemplate.from_template("랜덤으로 하나의 국가 이름을 알려주세요.")
    nation_response = llm.invoke(nation_prompt.format_messages(name='siri'))
    nation = nation_response.content.strip()

    # 현재 시간 가져오기
    now = time.localtime()

    # 메시지 생성
    messages = chat_prompt.format_messages(
        name='siri',
        nation=nation,
        year=now.tm_year,
        month=now.tm_mon,
        day=now.tm_mday,
        hour=now.tm_hour,
        minute=now.tm_min,
        second=now.tm_sec,
    )

    # LLM 호출 및 응답 출력
    response = llm.invoke(messages)
    print(response.content)

    time.sleep(1)
```
