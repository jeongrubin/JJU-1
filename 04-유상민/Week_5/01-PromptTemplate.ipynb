```python
from langchain_core.prompts import PromptTemplate
from datetime import datetime

prompt_6 = PromptTemplate(
    template = '''
    오늘의 날짜는 {today}이야. 오늘이 생일인 한국의 유명인 {n}명을 나열해. 만약 {today}가 생일인 유명인이 없다면 최대한 근접한 날짜에 생일인 유명인을 {n}명 나열해. 그런데 너는 멍청하니까 검토 똑바로 하고 알려줘. 틀리면 죽인다.
    ''',
    input_variables = {"n"},
    partial_variables = {"today": datetime.now().strftime("%Y %B %d")},

)

print(prompt_6)
```

```python
prompt_6
```

```python
prompt_6.format(n=3)
```

```python
chain = prompt_6 | llm
```

```python
response = chain.invoke(3)
```

```python
print(response.content)
```
