## DatFrameLoader

- pandas to docs

```python
import pandas as pd

df = pd.read_excel('/content/data/titanic.xlsx')

```

```python
from langchain.document_loaders import DataFrameLoader

loader = DataFrameLoader(df, page_content_column="name")

docs = loader.load()

print(docs[0].page_content)

print(docs[0].metadata)
```
