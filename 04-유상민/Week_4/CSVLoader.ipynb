### Export XML

```python
row = docs[1].page_content.split("\n")
row_str = "<row>"
for element in row:
    splitted_element = element.split(":")
    value = splitted_element[-1]
    col = ",".join(splitted_element[:-1])
    row_str += f"<{col}>{value.strip()}</{col}>"
row_str += "</row>"
print(row_str)
```

# Webbase loader

```python
!pip install -qU langchain_community
```

```python
import bs4
from langchain_community.document_loaders import WebBaseLoader

# WebBaseLoader 초기화
loader = WebBaseLoader(
    web_paths=['https://n.news.naver.com/article/029/0002927273?cds=news_media_pc&type=editn'],  # 올바른 키워드로 수정
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            "div",
            attrs={"class": ['newsct_article _article_body', 'go_trans _article_content']},
        )
    ),
    header_template={
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
)

# 문서 로드
docs = loader.load()

# 문서의 수 확인 및 출력
print(f"문서의 수: {len(docs)}")
print(docs)
```

```python
import bs4
from langchain_community.document_loaders import WebBaseLoader

# WebBaseLoader 초기화
loader = WebBaseLoader(
    web_paths=['https://smartstore.naver.com/toptime/products/6345008331'],
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            ["h3", "div"],
            attrs={"class": ["product_title__1XYZ", "product_detail__2ABC"]},
        )
    ),
    header_template={
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
)

# 문서 로드
docs = loader.load()

# 문서의 수 확인 및 출력
print(f"문서의 수: {len(docs)}")
print(docs)
```
