ğŸ§™ğŸ¾â€â™‚ï¸: ìœ„ ì½”ë“œë¥¼ ë³´ë©´, CSV íŒŒì¼ì„ ì½ì–´ì˜¬ ë•Œ delimiter, quotechar, ê·¸ë¦¬ê³  fieldnames ë“±ì„ ì§€ì •í•´ ì£¼ê³  ìˆì£ . ì™œ ì´ë ‡ê²Œ í•˜ëŠ” ê±¸ê¹Œìš”? ì£¼ëœ ì´ìœ ëŠ” CSV íŒŒì¼ì˜ êµ¬ì¡°ë¥¼ ëª…í™•í•˜ê²Œ ì¸ì‹í•˜ê³ , í•„ë“œ ì´ë¦„ì„ í™•ì‹¤íˆ ì§€ì •í•´ì£¼ê¸° ìœ„í•´ì„œì˜ˆìš”. íƒ€ì´íƒ€ë‹‰ ë°ì´í„°ì…‹ì˜ ê²½ìš°, ì›ë³¸ CSVì— ì´ë¯¸ í—¤ë”(ì—´ ì´ë¦„)ê°€ ìˆì„ ìˆ˜ë„ ìˆì§€ë§Œ, í•™ìŠµì´ë‚˜ ì „ì²˜ë¦¬ ê³¼ì •ì—ì„œ ì§ì ‘ ì—´ ì´ë¦„ì„ ì¬ì •ì˜í•´ì„œ ë‹¤ë£¨ê³  ì‹¶ì„ ë•Œê°€ ë§ê±°ë“ ìš”.

```python
# ì»¬ëŸ¼ì •ë³´:
# PassengerId,Survived,Pclass,Name,Sex,Age,SibSp,Parch,Ticket,Fare,Cabin,Embarked

# CSV íŒŒì¼ ê²½ë¡œ
loader = CSVLoader(
    file_path="/content/titanic.csv",
    csv_args={
        "delimiter": ",",  # êµ¬ë¶„ì
        "quotechar": '"',  # ì¸ìš© ë¶€í˜¸ ë¬¸ì
        "fieldnames": [
            "Passenger ID",
            "Survival (1: Survived, 0: Died)",
            "Passenger Class",
            "Name",
            "Sex",
            "Age",
            "Number of Siblings/Spouses Aboard",
            "Number of Parents/Children Aboard",
            "Ticket Number",
            "Fare",
            "Cabin",
            "Port of Embarkation",
        ],  # í•„ë“œ ì´ë¦„
    },
)

# ë°ì´í„° ë¡œë“œ
docs = loader.load()

# ë°ì´í„° ì¶œë ¥
print(docs[1].page_content)
```

```python
docs[1].metadata
```

ë¬¸ì„œ ì „ì²´ë¥¼ XML ë¬¸ì„œ í˜•ì‹ìœ¼ë¡œ ì²˜ë¦¬í•˜ë ¤ëŠ” ê²½ìš°
- ì°¸ê³ : 0ë²ˆì§¸ ë¬¸ì„œëŠ” í—¤ë” ì •ë³´ì´ê¸° ë•Œë¬¸ì— ìŠ¤í‚µí•©ë‹ˆë‹¤.

```python
row = docs[1].page_content.split("\n")
row_str = "<row>"
for element in row:
    splitted_element = element.split(":")
    value = splitted_element[-1]
    col = ":".join(splitted_element[:-1])
    row_str += f"<{col}>{value.strip()}</{col}>"
row_str += "</row>"
print(row_str)
```

Jun-----
- ìœ„ ë‚´ìš© ì „ì²´ ì¶œë ¥

```python
for doc in docs[1:]:
    row = doc.page_content.split("\n")
    row_str = "<row>"
    for element in row:
        splitted_element = element.split(":")
        value = splitted_element[-1]
        col = ":".join(splitted_element[:-1])
        row_str += f"<{col}>{value.strip()}</{col}>"
    row_str += "</row>"
    print(row_str)
```

`source_column` ì¸ìë¥¼ ì‚¬ìš©í•˜ì—¬ ê° í–‰ì—ì„œ ìƒì„±ëœ ë¬¸ì„œì˜ ì¶œì²˜ë¥¼ ì§€ì •í•˜ì„¸ìš”. ê·¸ë ‡ì§€ ì•Šìœ¼ë©´ ëª¨ë“  ë¬¸ì„œì˜ ì¶œì²˜ë¡œ `file_path`ê°€ ì‚¬ìš©ë©ë‹ˆë‹¤.

ì´ëŠ” CSV íŒŒì¼ì—ì„œ ë¡œë“œëœ ë¬¸ì„œë¥¼ ì¶œì²˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ì§ˆë¬¸ì— ë‹µí•˜ëŠ” ì²´ì¸ì— ì‚¬ìš©í•  ë•Œ ìœ ìš©í•©ë‹ˆë‹¤.

```python
loader = CSVLoader(
    file_path="/content/titanic.csv", source_column="PassengerId"
)  # CSV ë¡œë” ì„¤ì •, íŒŒì¼ ê²½ë¡œ ë° ì†ŒìŠ¤ ì»¬ëŸ¼ ì§€ì •

docs = loader.load()  # ë°ì´í„° ë¡œë“œ

print(docs[0])  # ë°ì´í„° ì¶œë ¥
```
