# SentenceTransformers



```python
from langchain.text_splitter import SentenceTransformersTokenTextSplitter

splitter = SentenceTransformersTokenTextSplitter(chunk_size = 200, chunk_overlap = 0)
```

```python
# 파일 객체 생성 및 내용을 읽어서 변수에 저장
with open('/content/appendix-keywords.txt') as f:
  file = f.read()

print(file[:350])
```

```python
count_start_and_stop_tokens = 2

# 텍스트 토큰 개수에서 시작과 종료 토큰의 개수를 센다
text_token_count = splitter.count_tokens(text = file) - count_start_and_stop_tokens
print(text_token_count)
```

```python
text_chunks = splitter.split_text(text = file) # 텍스트를 청크로 분할
print('0번째 청크:', text_chunks[0], '\n')
print('1번째 청크:', text_chunks[1])
```

- Huggging Face 토크나이저에 의해 계산된 토큰 수를 기준으로 함

```python
from transformers import GPT2TokenizerFast

hf_tokenizer = GPT2TokenizerFast.from_pretrained('gpt2')
```

```python
# 파일 객체 생성 및 내용을 읽어서 변수에 저장
with open('/content/appendix-keywords.txt') as f:
  file = f.read()

print(file[:350])
```

```python
text_splitter = CharacterTextSplitter.from_huggingface_tokenizer(
    hf_tokenizer,
    chunk_size = 300,
    chunk_overlap = 50
)

texts = text_splitter.split_text(file)
```

```python
print(texts[0]) # text 리스트의 1번째 요소 확인
```
