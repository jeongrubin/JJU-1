```python
from langchain.text_splitter import CharacterTextSplitter

# 텍스트를 분할할 때 사용할 구분자를 지정하고, 적당한 청크 크기로 설정
text_splitter = CharacterTextSplitter(
    separator="\n\n",  # 줄바꿈을 기준으로 분할
    chunk_size=100,  # 각 청크의 최대 크기 100자로 설정
    chunk_overlap=0,  # 중복 텍스트 없음
    length_function=len  # 길이를 계산하는 함수
)

# 텍스트를 분할하고 결과 출력
texts = text_splitter.create_documents([file])

# 텍스트 청크의 개수와 첫 번째 청크 출력
print(len(texts))  # 생성된 텍스트 청크의 개수
print(texts[0].page_content)  # 첫 번째 청크의 내용 확인
```

```python
texts = text_splitter.create_documents([file])
print(len(texts[0].page_content))
print(texts[0])
```

```python
# 문서에 대한 메타데이터 리스트를 정의
metadatas = [
    {'document' : 1},
    {'document' : 2},
    {'document' : 3},
]

# 분할할 텍스트 데이터를 리스트로 전달
documents = text_splitter.create_documents(
    [
        file,
        file,
        file,
    ],
    # 각 문서에 해당하는 메타데이터 전달
    metadatas = metadatas,
)

# 분할된 문서 중 첫 번째 문서를 출력
print(documents[0])
```
