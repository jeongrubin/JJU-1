# Python

````python
RecursiveCharacterTextSplitter.get_separators_for_language(Language.PYTHON)
````

````python
PYTHON_CODE = """
def open_data(df_path):
    df = pd.read_csv(df_path)
    filtered_data = df[(df['fractured'] == 1)]
    # ê²°ê³¼ ë°ì´í„°í”„ë ˆì„ ì¶œë ¥
    return filtered_data

# ì´ë¯¸ì§€ íŒŒì¼ì„ ì½ê³  ê·¸ë ˆì´ìŠ¤ì¼€ì¼ë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜
def open_image(image_dir, image_file):
    image_path = os.path.join(image_dir, image_file)  # ì´ë¯¸ì§€ íŒŒì¼ ê²½ë¡œ ìƒì„±
    return cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2GRAY)  # ì´ë¯¸ì§€ë¥¼ ê·¸ë ˆì´ìŠ¤ì¼€ì¼ë¡œ ë³€í™˜í•˜ì—¬ ë°˜í™˜

# JSON íŒŒì¼ì„ ì—´ê³  ë°ì´í„°ë¥¼ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜
def open_json(json_path):
    with open(json_path, "r") as f:
        json_data = json.load(f)  # JSON íŒŒì¼ì„ ì—´ê³  ë‚´ìš©ì„ ì½ì–´ì„œ ë°˜í™˜
    return json_data

def open_annotation(annotation_path):
    with open(annotation_path, 'r') as f:
        annotations = f.readlines()
    annotations = [line.strip().split() for line in annotations]
    return annotations
"""

python_splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.PYTHON, chunk_size=150, chunk_overlap=0
)
````

````python
python_docs = python_splitter.create_documents([PYTHON_CODE])
python_docs
````

````python
for i in python_docs:
  print(i.page_content)
  print('===' * 20)
````
